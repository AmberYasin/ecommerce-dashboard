```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

print("Downloading UCI Online Shoppers Intention dataset...")
dataset = fetch_ucirepo(id=468)

# Combine features and target
df = dataset.data.features.copy()
df['Revenue'] = dataset.data.targets

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 3 rows:")
print(df.head(3))
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
print("\nRevenue distribution:")
print(df['Revenue'].value_counts())
```

```python
import os
os.makedirs('../data/raw', exist_ok=True)
```

```python
# Validate row count
assert len(df) == 12330, f"Expected 12330 rows, got {len(df)}"
print(f"Row count confirmed: {len(df):,} rows")

# Check conversion rate
conversion_rate = df['Revenue'].mean() * 100
print(f"Conversion rate: {conversion_rate:.1f}%")

# Check visitor types
print(f"\nVisitor types:")
print(df['VisitorType'].value_counts())

# Check traffic types
print(f"\nUnique traffic types: {df['TrafficType'].nunique()}")

# Save as CSV
df.to_csv('../data/raw/uci_shoppers.csv', index=False)
print(f"\nSaved to data/raw/uci_shoppers.csv")
```
