```python
import pandas as pd
import os
```

```python
file_path = '/Users/itsupport/Documents/ecommerce-dashboard/data/raw/amazon_purchases.CSV'


df = pd.read_csv(file_path)
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 3 rows:")
print(df.head(3))
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Validate
assert len(df) > 100000, "Row count too low — check the file"
print(f"Row count confirmed: {len(df):,} rows")

# Check date range
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
print(f"Date range: {df['Order Date'].min()} to {df['Order Date'].max()}")

# Check categories
print(f"Unique categories: {df['Category'].nunique()}")

# Save as parquet — much faster to load in later stages
df.to_parquet('/Users/itsupport/Documents/ecommerce-dashboard/data/raw/amazon_purchases.parquet', index=False)
print("Saved parquet checkpoint successfully")
```
